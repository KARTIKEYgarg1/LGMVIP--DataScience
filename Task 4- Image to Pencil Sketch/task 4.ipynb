# Image to sketch

## Viewing file

```python
import cv2  
import matplotlib.pyplot as plt
img = cv2.imread('./flower.png')
img=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)  
```

## coverting image to grayscale

```python
gimg = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gimg)
```

## inverting grayscale image

```python
iimg = 255 - gimg
plt.imshow(iimg)
```

## Blurring inverted image

```python
blur = cv2.GaussianBlur(iimg,(7,7),0)
plt.imshow(blur)
```

## Proccesing images and producing sketch

```python
blur = 255 - blur
out = cv2.divide(gimg, blur, scale=256.0)
cv2.imshow("image",out)
cv2.waitKey()
```
